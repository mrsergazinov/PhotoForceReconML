#### Loading libraries

```python
# data processing and plotting libraries
import numpy as np
import os
import matplotlib.pylab as plt
import pandas as pd
import seaborn as sns
```

#### Define test errors for angle models

```python
#--- Test errors for contact angle ---#
df_error_inner = pd.DataFrame({"InceptionResNetV2" : np.array([0.0262, 0.1003, 0.1748, 0.2309, 0.1839]),
                              "VGG19" : np.array([0.0315, 0.0529, 0.0533, 0.0468, 0.0527]),
                              "Xception" : np.array([0.0704, 0.1561, 0.2187, 0.2182, 0.2145]), 
                              "Number of forces" : np.array(range(2, 7))})
df_error_inner = pd.melt(df_error_inner, id_vars=["Number of forces"], value_name="Mean absolute error", var_name="Model")
#--- Test errors for tangential angle ---#
df_error_tang = pd.DataFrame({"InceptionResNetV2" : np.array([0.0167, 0.0930, 0.1273, 0.2188, 0.2250]),
                              "VGG19" : np.array([0.0220, 0.0391, 0.2325, 0.2190, 0.2227]),
                              "Xception" : np.array([0.0307, 0.1040, 0.1632, 0.1705, 0.1777]), 
                              "Number of forces" : np.array(range(2, 7))})
df_error_tang = pd.melt(df_error_tang, id_vars=["Number of forces"], value_name="Mean absolute error", var_name="Model")
```

```python
sns.set_theme(style="whitegrid")
plt1 = sns.lineplot(data=df_error_inner, x="Number of forces", y="Mean absolute error", hue="Model")
plt1.set_xticks(range(2, 7))
plt1
```

```python
plt2 = sns.lineplot(data=df_error_tang, x="Number of forces", y="Mean absolute error", hue="Model")
plt2.set_xticks(range(2, 7))
plt2
```
